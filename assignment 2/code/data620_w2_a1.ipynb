```python
# video link 
# https://www.youtube.com/watch?v=elC2l7XVpqk

import networkx as nx
import matplotlib.pyplot as plt
import urllib

# Step 1: Load graph from a text file URL
url = 'https://raw.githubusercontent.com/Naik-Khyati/data620_web_anal/main/assignment%202/data/karate.txt'
response = urllib.request.urlopen(url)
G = nx.read_edgelist(response, nodetype=int)

# Step 2: Perform basic analysis
diameter = nx.diameter(G)  # Calculate the graph's diameter
average_clustering = nx.average_clustering(G)  # Calculate average clustering coefficient

print("Graph's Diameter:", diameter)
print("Average Clustering Coefficient:", average_clustering)

# Step 3: Visualization
nx.draw(G, with_labels=True)
plt.show()
```
