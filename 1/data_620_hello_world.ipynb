```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
g = nx.Graph()

# Add nodes (or vertices) to the graph object
g.add_nodes_from(["Andre", "Beverly", "Ed", "Carol", "Diane", "Fernando", "Garth", "Heather", "Ike", "Jane"])

# Add edges (representing relationships) between nodes
g.add_edges_from([
    ("Andre", "Carol"), ("Andre", "Diane"), ("Andre", "Fernando"),("Andre","Beverly"),
    ("Beverly", "Diane"), ("Beverly", "Ed"), ("Beverly", "Garth"),
    ("Carol", "Diane"), ("Carol", "Fernando"),
    ("Diane", "Ed"), ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"), ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
])

# Draw the graph
nx.draw(g, with_labels=True)

# Display the graph
plt.show()
```
